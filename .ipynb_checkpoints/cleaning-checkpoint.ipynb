```python
import sqlalchemy
import psycopg2
from sqlalchemy import create_engine
engine = create_engine('postgresql://postgres:Peter@1947@localhost/VotingProject')
```

```python
analysis_df = pd.read_sql("analysis",engine)
analysis_df["county_code"] =  analysis_df["fips"]
analysis_df = analysis_df.rename(columns={"fips":"FIPS"})
analysis_df.head()
```

```python
land_area_df = pd.read_sql("land_area",engine)
land_area_df.head()
```

```python
metro_df = pd.read_sql("metro",engine)
metro_df.head()
```

ll six tables

```python
pres_hist_df = pd.read_sql("pres_hist",engine)
pres_hist_df = pres_hist_df.rename(columns={"fips":"FIPS"})
pres_hist_df.head()
```

```python
states_df = pd.read_sql("states",engine)
states_df = states_df.rename(columns={"state_abbrev":"state_abbr"})
states_df = states_df.rename(columns={"state_name":"state"})
states_df.head()
```

```python
electoral_votes_df = pd.read_sql("electoral_votes",engine)
electoral_votes_df.head()
```

### combine the two state dataframes

```python
merge_df1 = pd.merge(states_df, electoral_votes_df, on="state_abbr",
    how="left")
merge_df1 =  merge_df1.set_index("state_abbr")
merge_df1.head()
```

### combine three dataframes by county code

```python
combined_df = pd.merge(analysis_df, land_area_df, on="FIPS",
    how="left")
combined_df =  combined_df.set_index("FIPS")
combined_df.head()
```

```python
combined_df = pd.merge(combined_df, metro_df, on="FIPS",
    how="left")
combined_df =  combined_df.set_index("FIPS")
combined_df.head()
```

### create separate tables for each year of presidential election history

```python
pres_hist_2000 =   pres_hist_df.loc[pres_hist_df["year"] == 2000,\
    ["FIPS", "dem", "rep", "oth", "tot"]]
pres_hist_2000 = pres_hist_2000.rename(columns={"dem":"dem2000"})
pres_hist_2000 = pres_hist_2000.rename(columns={"rep":"rep2000"})
pres_hist_2000 = pres_hist_2000.rename(columns={"oth":"oth2000"})
pres_hist_2000 = pres_hist_2000.rename(columns={"tot":"tot2000"})
pres_hist_2000.head()
```

```python
pres_hist_2004 =   pres_hist_df.loc[pres_hist_df["year"] == 2004,\
    ["FIPS", "dem", "rep", "oth", "tot"]]
pres_hist_2004 = pres_hist_2004.rename(columns={"dem":"dem2004"})
pres_hist_2004 = pres_hist_2004.rename(columns={"rep":"rep2004"})
pres_hist_2004 = pres_hist_2004.rename(columns={"oth":"oth2004"})
pres_hist_2004 = pres_hist_2004.rename(columns={"tot":"tot2004"})
pres_hist_2004.head()
```

```python
pres_hist_2008 =   pres_hist_df.loc[pres_hist_df["year"] == 2008,\
    ["FIPS", "dem", "rep", "oth", "tot"]]
pres_hist_2008 = pres_hist_2008.rename(columns={"dem":"dem2008"})
pres_hist_2008 = pres_hist_2008.rename(columns={"rep":"rep2008"})
pres_hist_2008 = pres_hist_2008.rename(columns={"oth":"oth2008"})
pres_hist_2008 = pres_hist_2008.rename(columns={"tot":"tot2008"})
pres_hist_2008.head()
```

```python
pres_hist_2012 =   pres_hist_df.loc[pres_hist_df["year"] == 2012,\
    ["FIPS", "dem", "rep", "oth", "tot"]]
pres_hist_2012 = pres_hist_2012.rename(columns={"dem":"dem2012"})
pres_hist_2012 = pres_hist_2012.rename(columns={"rep":"rep2012"})
pres_hist_2012 = pres_hist_2012.rename(columns={"oth":"oth2012"})
pres_hist_2012 = pres_hist_2012.rename(columns={"tot":"tot2012"})
pres_hist_2012.head()
```

```python
pres_hist_2016 =   pres_hist_df.loc[pres_hist_df["year"] == 2016,\
    ["FIPS", "dem", "rep", "oth", "tot"]]
pres_hist_2016 = pres_hist_2016.rename(columns={"dem":"dem2016"})
pres_hist_2016 = pres_hist_2016.rename(columns={"rep":"rep2016"})
pres_hist_2016 = pres_hist_2016.rename(columns={"oth":"oth2016"})
pres_hist_2016 = pres_hist_2016.rename(columns={"tot":"tot2016"})
pres_hist_2016.head()
```

### combine the three dataframes by county code 
### with the five presidential history dataframes

```python
combined_df = pd.merge(combined_df, pres_hist_2000, on="FIPS",
    how="left")
combined_df =  combined_df.set_index("FIPS")
combined_df.head()
```

```python
combined_df = pd.merge(combined_df, pres_hist_2004, on="FIPS",
    how="left")
combined_df =  combined_df.set_index("FIPS")
combined_df.head()
```

```python
combined_df = pd.merge(combined_df, pres_hist_2008, on="FIPS",
    how="left")
combined_df =  combined_df.set_index("FIPS")
combined_df.head()
```

```python
combined_df = pd.merge(combined_df, pres_hist_2012, on="FIPS",
    how="left")
combined_df =  combined_df.set_index("FIPS")
combined_df.head()
```

```python
combined_df = pd.merge(combined_df, pres_hist_2016, on="FIPS",
    how="left")
combined_df =  combined_df.set_index("FIPS")
combined_df.head()
```

### combine the three dataframes by county code

```python
combined_df = pd.merge(combined_df, merge_df1, on="state",
    how="left")
combined_df.head()
```
