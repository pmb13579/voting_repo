```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "cbsa-est2018-alldata1.csv"
df = pd.read_csv(csv_file)
df.head(20)
```

```python
MSA_df = df.loc[(df["LSAD"] == "Metropolitan Statistical Area") & (df["CENSUS2010POP"]>=1000000), ["CBSA", "NAME", "CENSUS2010POP"]]
MSA_df
```

```python
county_df = df.loc[df["LSAD"] == "County or equivalent", ["CBSA", "STCOU", "NAME", "CENSUS2010POP"]]
county_df.head()
```

```python
merge_table = pd.merge(MSA_df, county_df, on="CBSA",
    how="left")
merge_table.head()
```

```python
df1 = merge_table[["STCOU", "CBSA", "NAME_x" ]]
df2 = df1.rename(columns={"STCOU":"FIPS"})
df2 =  df2.set_index("FIPS")
df2.head()
```

```python
import sqlalchemy
import psycopg2
from sqlalchemy import create_engine
engine = create_engine('postgresql://postgres:Peter@1947@localhost/VotingProject')
df2.to_sql("metro",engine)
```
